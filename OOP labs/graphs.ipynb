```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_create_331(table, ax, show_sen=False):
    data = pd.read_csv(table, sep=';')
    if show_sen:
        sns.scatterplot(data=data, x='x', y='f_main', alpha=0.5, label='Основное SEN', s=10, c='blue', ax=ax)
        
    sns.scatterplot(data=data, x='x', y='f_mix', alpha=0.7, label='Смесь SEN', s=10, c='orange', ax=ax)
    sns.scatterplot(data=data, x='x', y='f_emp', alpha=0.5, label='Эмпирическая', s=10, c='green', ax=ax)
    ax.set_xlim(-9, 13)
    ax.set_xlabel('Значение случайной величины')
    ax.set_ylabel('Плотность')
    ax.legend(loc='upper right')
    ax.set_title(f'n={table.split("sen")[0]}')
    ax.grid(True, alpha=0.3)

def plot_create_332(table1, table2, ax):
    data1 = pd.read_csv(table1, sep=';')
    data2 = pd.read_csv(table2, sep=';')
    sns.scatterplot(data=data1, x='x', y='f_mix', alpha=0.7, label='Смесь SEN', s=10, c='orange', ax=ax)
    sns.scatterplot(data=data1, x='x', y='f_emp', alpha=0.5, label=f'Эмпирическая\n(из теста 3.3.1)', c='green', s=10, ax=ax)
    sns.scatterplot(data=data2, x='x', y='f_emp', alpha=0.5, label='Эмпирическая\n(новая)', s=10, c='red', ax=ax)
    ax.set_xlim(-9, 13)
    ax.set_xlabel('Значение случайной величины')
    ax.set_ylabel('Плотность')
    plt.legend(loc='upper right')
    ax.set_title('Два эмпирических n=1000')
    ax.grid(True, alpha=0.3)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15, 12), dpi=200)

plot_create_331('200sen_density_data.csv', axs[0, 0], show_sen=True)
plot_create_331('1000sen_density_data.csv', axs[0, 1])
plot_create_331('5000sen_density_data.csv', axs[0, 2])
plot_create_331('20000sen_density_data.csv', axs[1, 0])
plot_create_331('100000sen_density_data.csv', axs[1, 1])
plot_create_332('1000sen_density_data.csv', '1000_newsen_density_data.csv', axs[1, 2])

plt.tight_layout()
plt.suptitle("Сравнение плотностей распределений", y=1.01, fontsize='xx-large')
```
