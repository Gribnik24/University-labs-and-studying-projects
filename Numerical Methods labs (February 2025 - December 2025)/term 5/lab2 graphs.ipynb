```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def my_approx(x):
    if x >= -0.1 and x <= 0.3:
        return (40643 + 44510 * x) / 40000
    if x > 0.3 and x <= 0.6:
        return (8777 + 15740 * x) / 10000


def deepseek_approx(x):
    return 0.996307 + x * 0.980964 + x**2 * 0.65893
```

```python
x = np.linspace(-0.1, 0.6, 100)
ex = np.exp(x)
approx1 = np.vectorize(my_approx)(x)
approx2 = np.vectorize(deepseek_approx)(x)
```

```python
plt.figure(figsize=(10, 10))

sns.scatterplot(x=x, y=ex, label='Исходная функция', marker='X', alpha=0.8)
sns.scatterplot(x=x, y=approx1, label='Моя аппроксимация', alpha=0.5)
#sns.scatterplot(x=x, y=approx2, label='Аппроксимация DeepSeek', alpha=0.5)


plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
```
