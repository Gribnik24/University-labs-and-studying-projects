## 5. Выбор модели

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
from catboost import CatBoostRegressor, Pool

from sklearn.model_selection import GridSearchCV
```

```python

model_1 = LinearRegression()

model_2 = DecisionTreeRegressor(
    max_depth=7,      # Максимальная глубина дерева
    min_samples_split=2,  # Минимальное число образцов для разделения узла
    random_state=42
)

# Параметры для перебора
param_grid = {
    'depth': [4],               # Глубина деревьев
    'learning_rate': [0.01, 0.05], # Скорость обучения
    'iterations': [120],        # Число деревьев
    'l2_leaf_reg': [1],         # L2-регуляризация
}

model_3 = CatBoostRegressor(
                         loss_function='RMSE',
                         eval_metric='R2',
                         random_seed=42,
                         verbose = 1)  

# Настройка GridSearchCV
grid = GridSearchCV(
    estimator=model_3,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',  # Метрика (MSE)
    cv=3,                             # Количество фолдов
    n_jobs=-1                         # Параллельные вычисления
)
```

- Метрики регрессии:

    - MAE (Mean Absolute Error)
    - R² (коэффициент детерминации)

- Проверка на переобучение (разница в метриках на train/test).

```python
# Вычисляем RMSE LinearRegression
y_pred = model.predict(X_test)  # Предсказываем значения
rmse = np.sqrt(np.mean((y_test - y_pred) ** 2))

print(f"RMSE: {rmse:.4f}")
```
